```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define known points
x0, y0, z0 = [-0., -0.25,  0.0005]       # Start point
x1, y1, z1 = [0.015, -0.276,  0.0005]     # End point
h_clearance = 0.03           # Vertical clearance for y apex
d_deviation = 0.03           # Lateral deviation for z apex

# Apex points (midway)
x_apex = (x0 + x1) / 2
y_apex = max(y0, y1) + h_clearance
z_apex = max(z0, z1) + d_deviation

# Solve for parabolic coefficients for y(t) = a_y * t^2 + b_y * t + c_y
A_y = np.array([
    [0**2, 0, 1],  # y(0) = y0
    [0.5**2, 0.5, 1],  # y(0.5) = y_apex
    [1**2, 1, 1]   # y(1) = y1
])
b_y = np.array([y0, y_apex, y1])
a_y, b_y, c_y = np.linalg.solve(A_y, b_y)

# Solve for parabolic coefficients for z(t) = a_z * t^2 + b_z * t + c_z
A_z = np.array([
    [0**2, 0, 1],  # z(0) = z0
    [0.5**2, 0.5, 1],  # z(0.5) = z_apex
    [1**2, 1, 1]   # z(1) = z1
])
b_z = np.array([z0, z_apex, z1])
a_z, b_z, c_z = np.linalg.solve(A_z, b_z)

# Parameterize t
t = np.linspace(0, 1, 100)

# Generate 3D trajectory
x_t = (1 - t) * x0 + t * x1  # Linear interpolation for x
y_t = a_y * t**2 + b_y * t + c_y
z_t = a_z * t**2 + b_z * t + c_z

# Plot the trajectory
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_t, y_t, z_t, label="Foot Trajectory", lw=2)
ax.scatter([x0, x1], [y0, y1], [z0, z1], c='r', label="Start/End Points")
ax.scatter([x_apex], [y_apex], [z_apex], c='g', label="Apex Point")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.legend()
ax.set_title("3D Foot Trajectory (Parabolic in Y and Z)")
plt.show()


x_figsize = 15
y_figsize = 5
# Create a time array based on the number of columns in x1 or x2
# time = np.arange(state.shape[0])

# Create a single figure with a 3-row, 2-column layout
fig, axs = plt.subplots(1, 3, figsize=(x_figsize, y_figsize))
# fig.suptitle(title)

axs[0].scatter(t, x_t, label='x body', color='blue')
axs[0].set_xlabel('time')
axs[0].set_ylabel('x body')
axs[0].legend()
axs[0].grid(True)

axs[1].scatter(t, y_t, label='y body', color='blue')
axs[1].set_xlabel('time')
axs[1].set_ylabel('y body')
axs[1].legend()
axs[1].grid(True)

axs[2].scatter(t, z_t, label='z body', color='blue')
axs[2].set_xlabel('time')
axs[2].set_ylabel('z body')
axs[2].legend()
axs[2].grid(True)
```

```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define known points
x0, y0, z0 = [-0., -0.25,  0.0005]       # Start point
x1, y1, z1 = [0.015, -0.276,  0.0005]     # End point
h_clearance = 0.03           # Vertical clearance for y apex
d_deviation = 0.03           # Lateral deviation for z apex

# Apex points (midway)
x_apex = (x0 + x1) / 2
y_mid = (y0 + y1) / 2
z_mid = (z0 + z1) / 2

y_apex = y_mid + h_clearance
z_apex = z_mid + d_deviation

# Solve for parabolic coefficients for y(t) = a_y * t^2 + b_y * t + c_y
A_y = np.array([
    [0**2, 0, 1],    # y(0) = y0
    [0.5**2, 0.5, 1],  # y(0.5) = y_apex
    [1**2, 1, 1]     # y(1) = y1
])
b_y = np.array([y0, y_apex, y1])
coeffs_y = np.linalg.solve(A_y, b_y)
a_y, b_y, c_y = coeffs_y

# Solve for parabolic coefficients for z(t) = a_z * t^2 + b_z * t + c_z
A_z = np.array([
    [0**2, 0, 1],    # z(0) = z0
    [0.5**2, 0.5, 1],  # z(0.5) = z_apex
    [1**2, 1, 1]     # z(1) = z1
])
b_z = np.array([z0, z_apex, z1])
coeffs_z = np.linalg.solve(A_z, b_z)
a_z, b_z, c_z = coeffs_z

# Parameterize t
t = np.linspace(0, 1, 100)

# Generate 3D trajectory
x_t = (1 - t) * x0 + t * x1  # Linear interpolation for x
y_t = a_y * t**2 + b_y * t + c_y
z_t = a_z * t**2 + b_z * t + c_z

# Plot the trajectory
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_t, y_t, z_t, label="Foot Trajectory", lw=2)
ax.scatter([x0, x1], [y0, y1], [z0, z1], c='r', label="Start/End Points")
ax.scatter([x_apex], [y_apex], [z_apex], c='g', label="Apex Point")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.legend()
ax.set_title("3D Foot Trajectory (Parabolic in Y and Z)")
plt.show()
```
